```python
import os
import matplotlib.pyplot as plt
import numpy as np
from scipy.io.wavfile import read
from scipy.io.wavfile import write
from IPython.display import Audio
from scipy import signal
%matplotlib inline
print(os.getcwd())
```

```python
sampling_rate, data = read('with_hum.wav')
Audio('with_hum.wav')
```

```python
sp = np.fft.fft(data)
freq = np.fft.fftfreq(len(data),1/sampling_rate)
plt.plot(freq,abs(sp))
plt.title('Assignment 1')
plt.ylabel('magnitude')
plt.xlabel('frequency')
plt.show()
```

```python
b,a = signal.butter(10,420,'highpass',fs=sampling_rate*0.5)
w,h=signal.freqs(b,a)
plt.plot(w,abs(h))
plt.title('filter frequency response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.figure()
plt.plot(w,np.angle(h))
plt.title('filter phase')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Phase [degree]')
print (sampling_rate)
```

```python
#The design has 10 taps, this is due to with lower taps, the 400HZ signal is not cleanly filtered out,
#If we raise the cut_off freq in the butter function, we would be able to reduce the taps needed,
#but we might lose useful signals too.
#with lower taps, the delay and resource utilization is less.
```

```python
data_out = signal.lfilter(b,a,data)
data_out_fft = np.fft.fft(data_out)
data_freq = np.fft.fftfreq(data_out_fft.size,1/sampling_rate)

plt.plot(data_freq,abs(data_out_fft))
plt.title('Assignment 3')
plt.ylabel('magnitude')
plt.xlabel('frequency')
plt.show()
print(abs(data_out))
print(abs(data))
```
